```python
import requests
import pandas as pd

# URL RAW del JSON
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"

# Obtener los datos desde la API
response = requests.get(url)
response.raise_for_status()  # Lanzar√° error si hay problemas
data = response.json()

# Mostrar tipo de dato base
print(" Tipo de datos:", type(data))
print(" Total registros:", len(data))

# Mostrar un registro de ejemplo
print("\n Ejemplo de registro:")
print(data[0])

# Convertir a DataFrame
df = pd.DataFrame(data)

# Mostrar las columnas y tipos
print("\n Columnas del DataFrame:")
print(df.dtypes)

# Mostrar primeras filas
df.head()
```
